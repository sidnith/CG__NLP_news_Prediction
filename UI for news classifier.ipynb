```python
import os
import string
import tkinter
from tkinter import *
import requests
url = 'http://localhost:8010/'
```

```python
def print_result(output):
    result.grid(row=7,columnspan=no_of_modes,padx=225)
    var.set('Predicted as : {} category article\n------------------------------------'.format(output[0]))
    #prediction_explainer_button.grid(row=8,columnspan=no_of_modes)
    #prediction_explainer_button.configure(command=prediction_explainer)
    recommendation_label_0.grid(row=9,columnspan=no_of_modes)
    recommendation_label_1.grid(row=10,columnspan=no_of_modes)
    recommendation_label_2.grid(row=11,columnspan=no_of_modes)
    recommendation_label_3.grid(row=12,columnspan=no_of_modes)
    recommendation_label_4.grid(row=13,columnspan=no_of_modes)
    recommendation_label_5.grid(row=14,columnspan=no_of_modes)
    var1_1.set("{} \t\t\t {} \t\t\t {}".format(output[2][0].rstrip(),output[3][0],output[4][0]))
    var1_2.set("{} \t\t\t {} \t\t\t {}".format(output[2][1].rstrip(),output[3][1],output[4][1]))
    var1_3.set("{} \t\t\t {} \t\t\t {}".format(output[2][2].rstrip(),output[3][2],output[4][2]))
    var1_4.set("{} \t\t\t {} \t\t\t {}".format(output[2][3].rstrip(),output[3][3],output[4][3]))
    var1_5.set("{} \t\t\t {} \t\t\t {}".format(output[2][4].rstrip(),output[3][4],output[4][4]))
```

```python

def hide_text_mode():
    L1.grid_forget()
    t.grid_forget()
    result.grid_forget()
    hide_recommendation_result()
def hide_batch_mode():
    L2.grid_forget()
    L3.grid_forget()
    e1.grid_forget()
    e2.grid_forget()
    result.grid_forget()
    hide_recommendation_result()
def hide_single_file_mode():
    L4.grid_forget()
    e3.grid_forget()
    result.grid_forget()
    single_file_text.grid_forget()
    hide_recommendation_result()
def hide_recommendation_result():
    recommendation_label_0.grid_forget()
    recommendation_label_1.grid_forget()
    recommendation_label_2.grid_forget()
    recommendation_label_3.grid_forget()
    recommendation_label_4.grid_forget()
    recommendation_label_5.grid_forget()
def show_text_mode():
    hide_batch_mode()
    hide_single_file_mode()
    L1.grid(row=2,padx=2)
    t.grid(row=3,columnspan=no_of_modes,padx=20)
    B.configure(command=text_mode)
def show_batch_mode():
    hide_text_mode()
    hide_single_file_mode()
    L2.grid(row=3,column=0)
    L3.grid(row=4,column=0)
    e1.grid(row=3,column=1,pady=30)
    e2.grid(row=4,column=1,pady=30)
    B.configure(command=batch_mode)
def show_single_file_mode():
    hide_text_mode()
    hide_batch_mode()
    L4.grid(row=3,column=0)
    e3.grid(row=3,column=1)
    B.configure(command=single_file_mode)
    
def show_selected_mode_input_field():
    B.grid(row=6, column=0,columnspan=no_of_modes,padx=250,pady=20)
    if v.get()==1:
        show_text_mode()
       
    if v.get()==2:
        show_batch_mode()
        
    if v.get()==3:
        show_single_file_mode()
```

```python
import os
def text_mode():
    text=t.get('1.0', END)
    output = requests.post(url+'api',json={'content':text})
    print(output.json())
    print_result(output.json())
    
def batch_mode():
    sourcePath=e1.get()
    destPath=e2.get()
          
    for files in os.walk(e1.get()):
        print(files[2])
    for i in files[2]:
        filepath=sourcePath+"/"+i 
        f = open(filepath, "r")
        text=f.read()
        output = requests.post(url+'api',json={'content':text})
        print(i,"--",output.json())
        file1 = open((destPath+"/" +output.json()+i), "w")
        file1.write(text)
        f.close()
        file1.close()
        
def single_file_mode():
    sourcePath=e3.get()
    f=open(sourcePath,'r')
    text=f.read()
    print(text)
    output = requests.post(url+'api',json={'content':text})
    print(output.json())
    print_result(output.json())
    single_file_text.grid(row=16,columnspan=3)
    var2.set('\n\nArticle Content:\n\n{}'.format(text))

def prediction_explainer():
    requests.post(url+'prediction_explainer')
    
    
```

```python
top = tkinter.Tk()
# Code to add widgets will go here...
top.title('News Classification')
screen_length=750
screen_height=750
no_of_modes=3
top.geometry("{}x{}".format(screen_length,screen_height))
top.resizable(0, 0) 


Label(top,text="Select Mode : ").grid(row=0,columnspan=2,padx=250)
v = tkinter.IntVar()
Radiobutton(top,text="Text mode",padx = 20, variable=v,value=1,command=show_selected_mode_input_field).grid(row=1,column=0)
Radiobutton(top,text="Batch mode",padx = 20, variable=v,value=2,command=show_selected_mode_input_field).grid(row=1,column=1)
Radiobutton(top,text="Single file mode",padx = 20, variable=v,value=3,command=show_selected_mode_input_field).grid(row=1,column=2)


B = tkinter.Button(top, text ="Classify")
prediction_explainer_button = tkinter.Button(top, text ="why?")
t = Text(top, height=20, width=63)
L1=Label(top, text="Enter text to be classified",justify='left')

L2=Label(top,text='Enter path of unclassified documents : ',justify='left')
L3=Label(top,text='Path to save classified documents : ',justify='left')
e1=Entry(top, width=50)
e2=Entry(top, width=50)

L4=Label(top,text='Enter path of file to predict : ',anchor='w')
e3=Entry(top,width=50)
var=tkinter.StringVar()
var1_1=tkinter.StringVar()
var1_2=tkinter.StringVar()
var1_3=tkinter.StringVar()
var1_4=tkinter.StringVar()
var1_5=tkinter.StringVar()

var2=tkinter.StringVar()
single_file_text=Label(top,textvariable=var2,wraplength=700)


result=Label(top, textvariable=var,foreground='red')
recommendation_label_0=Label(top,text='\n\n\t\tRecommended Similar Articles\n\n\t Headlines \t\t\t\t   Similarity score \t\t\t\t   filename',foreground='Green')
recommendation_label_1=Label(top,textvariable=var1_1,foreground='blue')
recommendation_label_2=Label(top,textvariable=var1_2)
recommendation_label_3=Label(top,textvariable=var1_3,foreground='blue')
recommendation_label_4=Label(top,textvariable=var1_4)
recommendation_label_5=Label(top,textvariable=var1_5,foreground='blue')

top.mainloop()
```
